```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('./datasets/user_behavior_dataset.csv')

# Basic info
print(df.head())  # First 5 rows
print(df.info())  # Data types, nulls
print(df.describe())  # Stats summary

# Check for missing values
print(df.isnull().sum())  # Should be zero

# Visualize correlations (focus on drain)
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlations')  
plt.show()

# Pairplot for key features vs. drain
sns.pairplot(df[['App Usage Time (min/day)', 'Screen On Time (hours/day)', 
                 'Data Usage (MB/day)', 'Battery Drain (mAh/day)']])
plt.show()

# Boxplot: Drain by OS
sns.boxplot(x='Operating System', y='Battery Drain (mAh/day)', data=df)
plt.title('Battery Drain by OS')
plt.show()
```
